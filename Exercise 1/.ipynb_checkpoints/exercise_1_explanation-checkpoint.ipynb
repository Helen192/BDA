Next write a scaler to scale the values of a column to an interval between a new minimum and maximum value.

To achieve this your scaler can implement the formula:  
$$\frac{max_{new}-min_{new}}{max_{old}-min_{old}}\cdot (v-min_{old})+min_{new}$$
where $max_{new}$ and $min_{new}$ are the new boundaries of the values, $max_{old}$ and $min_{old}$ are the old boundaries and $v$ is a value from the column that will be scaled.

_Hint: Do **not** use a loop over all entries of the column! What other ways are there to manipulate columnar data?_

````python
def my_scaler(column,min_val,max_val):
    scaled_column = (((max_val - min_val)/(column.max() - column.min()))
                     *(column - column.min())) + min_val
    return scaled_column
````

## Using vectorized functions
In order to avoid python loops, we make use of vectorized funtions provided by the Pandas library (they are methods of the `pandas.Series` object `column`). The above implementation uses overloaded operators for convenience and readability but note that our function executes three vectorized operations on the input parameter `column` of type `pandas.Series`.

First, we subtract a scalar value from all elements in the vector, implementing $v-min_{old}$:
```
shifted_to_zero = column.sub(column.min())
```
Next, we compute a normalization factor (which is a scalar value) and multiply our shifted vector by this factor, implementing $\frac{max_{new}-min_{new}}{max_{old}-min_{old}}\cdot (v-min_{old})$:
```
norm_scalar = (max_val - min_val)/(column.max() - column.min())
shifted_normalized = shifted_to_zero.mul(norm_scalar)
```
Finally, we shift all values back by $min_{new}$:
```
scaled_column = shifted_normalized.add(min_val)
```

> The only reason why you would explicitly call these methods is when you need to change the default arguments for one of the remaining parameters `fill_value` and `level` (see the respective method documentation, it is not needed here).

## Continuing exercise 1: scaling the values
* Now use your scaler on the column ``value_eur`` and scale it to an interval between $0$ and $100$.  
* Then take the new values for the ``value_eur`` und turn them into discrete values. If a player is below $25$ he should be categorized as ``low``, between $25$ and $50$ he should be categorized as ``low_average`` then between $50$ and $75$ as ``high_average`` and finally any that are above $75$ should be categorized as ``high``.  
* Finally create a histogram to show the distribution of the players' value.

````python
scaled_values = my_scaler(selected_df.loc[:,'value_eur'],0,100)
selected_df.loc[:,'value_eur'] = scaled_values
````

````python
bins = [0, 25, 50, 75, 101]
selected_df.loc[:,'value_eur'] = np.digitize(selected_df.loc[:,'value_eur'], bins)
````

## Why do we need to specify the last bin edge as $101$ even though we want to bin values up to $100$?
Try this:

````python
np.digitize(df.loc[:,'value_eur'], [0, 25, 50, 75, 100]).max()
````

The largest bin number is now $5$. Why $5$? because the condition the returned indices satisfy (see the documentation for the `np.digitize` function) is `bins[i-1] <= x < bins[i]` and since the value $100$ is not strictly smaller than any of the bins: 
$$100 \nless \mathrm{bins}[4]=100$$
Therefore, the function returns the next larger integer $4+1=5$ as the bin index. You may already guess what happens when we try to digitize a value that is smaller than the smallest bin edge:

````python
np.digitize([-1], [0, 25, 50, 75, 100])
````

## Continuing with exercise 1: mapping to categorical values

````python
selected_df.loc[:,'value_eur'] = np.array(['low', 'low_average', 'high_average',
                                     'high'])[selected_df.loc[:,'value_eur']-1]
````

## Why do we need to subtract $1$?
So provided that all values $x$ given as an input to the `np.digitize` function satisfy
$$0\le x \lt 101$$
they will all be mapped to integer bin indices $i_\mathrm{bin}\in[1,4]$. Now we want to use these indices to index an array with the categorical labels. Therefore, we need to subtract $1$ (remember that array indexing is zero-based!).

## Continuing with exercise 1: Drawing the histogram
To plot a histogram of the values in the `value_eur` column, we can use the convenience function `pandas.Series.hist` (a method of the Series object) with default parameters (see the [documentation](https://pandas.pydata.org/docs/reference/api/pandas.Series.hist.html?highlight=hist#pandas.Series.hist) for more) which uses matplotlib: 

````python
selected_df['value_eur'].hist();
````

## Why is the ordering of histogram bins from 'high' to 'low'?
The ordering of the categorical bins is random. Python does not have any concept of the semantic meaning of our labels. From python 3.6, the order of categorical values in axes by default depends on the order in which they are first encountered in the Dataframe. 

To see that this is the only reason for this ordering, let's draw the histogram for a view on our Dataframe that reverses the order of its index:

````python
selected_df['value_eur'][::-1].hist();
````

Let's turn to the `plotly` library to specify our own arbitrary ordering of the categorical labels. Also, let's switch to a logarithmic $y$-axis to better compare the values of the three less common categories:

````python
import plotly.express as px

fig = px.histogram(selected_df, x='value_eur', category_orders={'value_eur': ['low', 'high', 'high_average', 'low_average']}, log_y=True)
fig.show()
````

# Exercise 3

Group the players by the club they are playing for. Then calculate the average of the ``overall`` column for each team.  

````python
club_overalls = selected_df.loc[:,['club','overall']].groupby('club').mean()
````

Then print the five Teams with the highest as well as the five teams with the lowest average ``overall`` value, so we can see which teams are the strongest and which teams are the weakest.

````python
club_overalls.sort_values(by='overall', ascending=False)[:5]
````

````python
club_overalls.sort_values(by='overall', ascending=False)[-5:]
````

## How does indexing with negative numbers, i.e. `[-5:]` work?
Let's find the club with the second to fourth smallest `overall` values. We ordered the values from high to low, so they should be the fourth-to-last until the second-to-last elements. Since slices in python exlude the upper boundary, we need to write `[-4:-1]` for our slice of rows.

````python
club_overalls.sort_values(by='overall', ascending=False)[-4:-1]
````

Try setting the parameter `ascending=True`, what happens? How do we obtain our clubs now?

````python
club_overalls.sort_values(by='overall', ascending=True)[1:4]
````

## How can we use pandas' `.loc` method here?
Note that `club_overalls` is a `pandas.Dataframe`:

````python
type(club_overalls)
````

... that has only one column:

````python
club_overalls.columns
````

... and whose index is not numerical, but contains strings instead:

````python
club_overalls.index
````

Which is why using the `.loc` method with our integer slice will not work:

````python
try:
    club_overalls.sort_values(by='overall', ascending=True).loc[1:4, :]
except TypeError as e:
    print(e)
````

If we know the labels of the rows we are looking for, we can use label-based indexing like this. Note that we can either get all columns by giving the None-slice `:` as the second argument or specify `overall`, which is the only column anyways.

````python
club_overalls.sort_values(by='overall', ascending=True).loc['Finn Harps':'Waterford FC', :]

# what happens if you write the following line of code instead? Why? How can you fix this (apart from changing the order of labels back again)?
# club_overalls.sort_values(by='overall', ascending=True).loc['Waterford FC':'Finn Harps', :]
````

Since we generally do not know the answer to our query a-priori, we can also use a pandas function for integer-based indexing (see [selection by position](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#indexing-integer)). It is called `iloc`:

````python
club_overalls.sort_values(by='overall', ascending=True).iloc[1:4, :]
````

Note that the second argument to `.iloc[]` can no longer be `overall` (which is a label):

````python
try:
    club_overalls.sort_values(by='overall', ascending=True).iloc[1:4, 'overall']
except ValueError as e:
    print(e)
````

Instead, we have to give a slice or a list of the integer indices of the columns we want. Try omitting the list and just giving `0` as the second argument. Why is this output formatted differently? What type does the returned value have?

````python
club_overalls.sort_values(by='overall', ascending=True).iloc[1:4, [0]]
````

Now let's delete the call to the `copy()` method when creating the `selected_df`-Dataframe and run the notebook again. What has changed?

We get a `SettingWithCopy` warning from Pandas! What happened? The initial assignment of `selected_df` is ambiguous. Pandas cannot predict whether what has been assigned to this variable is a view or a copy of the original Dataframe. Consequently, it is unclear whether you modify a temporary copy or the original data. **That is what Pandas is warning you about at every following assignment to `selected_df`.**. See [indexing view versus copy](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#indexing-view-versus-copy) in the Pandas docs for more detail. On a sidenote, chained indexing may also be a little slower, because it does not allow Pandas to optimize.

To fix this, you can either call the `copy()` method again to explicitly demand a copy of the view (if it is one). Or you can use the `.loc` indexing function as well, which will return a copy. Don't forget to add the None-slice `:` to include all rows!

# When to use `.loc[]` versus python-based indexing `[]`
Using `[]` is more convenient but it can also introduce ambiguouties and is usually slower. You should generally use `.loc[]` (or `.iloc` if the former cannot be used for your application (see above for an example of that)), especially if your Dataframe has more than one column!
